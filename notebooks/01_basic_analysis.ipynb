# 01 – Basic Experiment Analysis

Load a run from MLflow, inspect its metadata, and plot a learning curve.

```python
from phinance.experiment.results import list_runs
from phinance.experiment.visualization import get_run_data, plot_learning_curve
import pandas as pd
```

## 1) List runs
Adjust `experiment_name` to an experiment that exists in your MLflow store.

```python
experiment_name = None  # e.g. 'execution_agent_test'
runs_df = list_runs(experiment_name=experiment_name)
runs_df.head()
```

## 2) Select a run

```python
if runs_df.empty:
    raise ValueError('No runs found. Execute an experiment first.')
run_id = runs_df.iloc[0]['run_id']
run_id
```

## 3) Load run data

```python
run_data = get_run_data(run_id)
pd.DataFrame([run_data['params']]).T.rename(columns={0: 'value'}).head()
```

## 4) Plot a learning curve

```python
metric_name = 'reward'
fig = plot_learning_curve([run_id], metric=metric_name)
fig
```
